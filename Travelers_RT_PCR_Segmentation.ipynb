By taking visualization part and WHO dataset under consideration Three priority levels(segments) has been defined and based on this segments the attention criteria can be fulfilled.

The WHO dataset, according to the country, was observed to have countries, cases - newly reported in the last 7 days, and cases - newly reported in the last 7 days per 100000 population. With these attributes, two new attributes were created: the percentage of cases - newly reported in the last 7 days per 100000 population  and the level of risk. 


**percenatge** = {X} / {sum(X)} * 100100 
where, X = cases newly reported in last 7 days per 100000 population


The three values were assigned to the level of risk according to the percentage of cases where,
\begin{align}
  risk level=\begin{cases}
    2, & \text{if $ percent > 2 $}\\
    1, & \text{if $$ percent < 2 <= 1 $$}\\
    0, & \text{$otherwise$}
  \end{cases}
\end{align}

The final step is to segment the international travellers in three possible segments - Immediate Priority, High Priority and Low Priority

*   **Immediate Priority**- The Passengers who need immediate attention, i.e. should be selected for RTPCR Testing

*   **High Priority and Medium Priority** - Passengers who can be randomly selected for RTPCR Testing, but donot require immediate attention

*  **Lower Priority** - Passengers that dont fulfil the criteria for Immediate or High Priority.

<br/> The Definition of the Priority is decided by Canada's Covid Cases Data vlsualizations from last 7 days based on the factors of Age and Gender

Based on the Data Visualizations, here are the definitions of the Segments

**Immediate Priority Level 3**- Travellers with Age 70 and above, RTPCR Postive Reports greater than 14 days (2 weeks) old. Both Males and Females, Travelling from  High affected countries (Percentage of Cases - newly reported in last 7 days per 100000 population is 2% and above; reference : WHO Data).

**Immediate Priority Level 2 A**- Travellers with Age 70 and above, RTPCR Postive Reports less than 14 days (2 weeks) old. Both Males and Females, Travelling from High affected countries (Percentage of Cases - newly reported in last 7 days per 100000 population is 2% and above; reference : WHO Data).

**Immediate Priority Level 2 B**- Travellers with Age 70 and above, RTPCR Postive Reports greater than 14 days (2 weeks) old. Both Males and Females, Travelling from low affected countries (Percentage of Cases - newly reported in last 7 days per 100000 population is 1% and below; reference : WHO Data).


**High Priority Level 3**- Travellers with Age 50 to 69, RTPCR Postive Reports greater than 14 days old. Both Males and Females, Travelling from High affected countries (Percentage of Cases - newly reported in last 7 days per 100000 population is 2% and above; reference : WHO Data). 

**High Priority Level 2 A**- Travellers with Age 50 to 69, RTPCR Postive Reports less than 14 days (2 weeks) old. Both Males and Females, Travelling from High affected countries (Percentage of Cases - newly reported in last 7 days per 100000 population is 2% and above; reference : WHO Data). 

**High Priority Level 2 B**- Travellers with Age 50 to 69, RTPCR Postive Reports greater than 14 days old. Both Males and Females, Travelling from low affected countries (Percentage of Cases - newly reported in last 7 days per 100000 population is 1% and below; reference : WHO Data).

**Medium Priority Level 2 A** : Travellers with Age 50 and above with RTPCR Negative Reports.

**Medium Priority Level 2 B** : Travellers with Age 0 to 49 with RTPCR Postive Reports

**Low Priority Level** : Travellers with Age 0 to 49 with RTPCR Negative Reports

# High level 2B

Travellers with Age 50 to 69, RTPCR Postive Reports greater than 14 days old. Both Males and Females, Travelling from low affected countries (Percentage of Cases - newly reported in last 7 days per 100000 population is 1% and below; reference : WHO Data).

# Medium 2A

Travellers with Age 50 and above with RTPCR Negative Reports.

```python
medium_2a = report_data[(report_data['Age'] > 50) & (report_data['PCR_Status'] == 'NO') & (report_data['RTPCR Validity'] > 14)]
```
